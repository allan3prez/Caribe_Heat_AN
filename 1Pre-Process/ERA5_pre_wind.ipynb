# find file paths

```python
years = np.arange(1981, 2027, 1)
years
```

```python
u_files = sorted(glob.glob(path+'*/*10u*.nc'))
u_files
```

```python
v_files = sorted(glob.glob(path+'*/*10v*.nc'))
v_files
```

# crop and get

```python
def crop_u_ds(ds):
    ds = ds['VAR_10U']
    ds.coords['longitude'] = (ds.coords['longitude'] + 180) % 360 - 180
    ds = ds.sortby(ds.longitude)
    ds = ds.sortby(ds.latitude)
    ds_crop = ds.sel(latitude=slice(7.75, 28), longitude=slice(-89, -59))
    ds_crop = ds_crop.sel(time=ds_crop.time.dt.year.isin(years))
    return ds_crop


U = xr.open_mfdataset(u_files, combine='nested', concat_dim='time',
                     engine='h5netcdf', parallel=True,
                     preprocess=crop_u_ds,
                     chunks={})
U
```

```python
U = U.chunk({'time': (U.sizes['time'] / 12), 'latitude': -1, 'longitude': -1})
U
```

```python
def crop_v_ds(ds):
    ds = ds['VAR_10V']
    ds.coords['longitude'] = (ds.coords['longitude'] + 180) % 360 - 180
    ds = ds.sortby(ds.longitude)
    ds = ds.sortby(ds.latitude)
    ds_crop = ds.sel(latitude=slice(7.75, 28), longitude=slice(-89, -59), time=ds.time.dt.year.isin(years))
    return ds_crop


V = xr.open_mfdataset(v_files, combine='nested', concat_dim='time',
                     engine='h5netcdf', parallel=True,
                     preprocess=crop_v_ds,
                     chunks={})
V = V.chunk({'time': (V.sizes['time'] / 12), 'latitude': -1, 'longitude': -1})
V
```
