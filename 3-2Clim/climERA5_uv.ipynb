```python
import xarray as xr
```

```python
path = '/glade/work/acruz/Caribbean_Heat_data/ERA5/'
u = xr.open_dataset(path+'U10_during_peak_HI.nc')
v = xr.open_dataset(path+'V10_during_peak_HI.nc')
```

```python
u_clim = u.groupby('time.month').mean()
u_clim = u_clim.rename({'U_during_HIdmax': 'U_HIdmax_clim'})
u_clim
```

```python
v_clim = v.groupby('time.month').mean()
v_clim = v_clim.rename({'V_during_HIdmax': 'V_HIdmax_clim'})
v_clim
```

```python
u_clim.to_netcdf(path+'U_HIdmax_clim.nc')
```

```python
v_clim.to_netcdf(path+'V_HIdmax_clim.nc')
```
