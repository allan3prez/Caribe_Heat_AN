```python
import xarray as xr
```

```python
path = '/glade/work/acruz/Caribbean_Heat_data/ERA5/'
```

```python
u_clim = xr.open_dataset(path+'U_HIdmax_clim.nc')
v_clim = xr.open_dataset(path+'V_HIdmax_clim.nc')
u = xr.open_dataset(path+'U10_during_peak_HI.nc')
v = xr.open_dataset(path+'V10_during_peak_HI.nc')
```

```python
u
```

```python
u_anom = u['U_during_HIdmax'].groupby('time.month') - u_clim
u_anom
```

```python
v_anom = v['V_during_HIdmax'].groupby('time.month') - v_clim
v_anom
```

```python
u_anom = u_anom.rename({'U_HIdmax_clim': 'U10_HIdmax_anom'})
v_anom = v_anom.rename({'V_HIdmax_clim': 'V10_HIdmax_anom'})
```

```python
u_anom.to_netcdf(path+'U_Hidmax_anom.nc')
```

```python
v_anom.to_netcdf(path+'V_Hidmax_anom.nc')
```
